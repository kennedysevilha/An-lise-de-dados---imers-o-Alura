```python
# Importaçoes
import pandas as pd
import plotly.express as px

# OBS: caso usar pelo google
# excel: google sheet     dica de extençao: sheetgpt
# vscode: google colab
```

```python
# --- ler tabelas
df_principal = pd.read_excel("tabela.xlsx" , sheet_name="Principal") # sheet_name selecionar aba do excel
df_principal.head(10) # mostra o intervalo de linhas, sem o head um mostra geral
```

```python
df_total_de_acoes = pd.read_excel("tabela.xlsx" , sheet_name="Total_de_acoes")
df_total_de_acoes.head()
```

```python
df_Ticker = pd.read_excel("tabela.xlsx" , sheet_name="Ticker")
df_Ticker.head() 
```

```python
df_Chat_GPT = pd.read_excel("tabela.xlsx" , sheet_name="Chat_GPT")
df_Chat_GPT
```

```python
# filtras colunas de uso
df_principal = df_principal[["Ativo", "Data", "Último (R$)","Var. Dia (%)"]].copy() # copiar tabelas/ redefinir 
df_principal
```

```python
# redefinindo nome das colunas/ evitar erro
df_principal = df_principal.rename(columns= {"Último (R$)": "valor_final", "Var. Dia (%)" : "var_dia_pct"}).copy() # renomear colunas { "nome atual" : "novo nome" }
df_principal
```

```python
# Criar nova coluna, pegando a anterios e dividindo 100
df_principal["Var_pct"] = df_principal["var_dia_pct"]/100
# Criando nova coluna para o valor inicial 
df_principal["valor_inicial"] = df_principal["valor_final"] /(1+ df_principal["Var_pct"])
df_principal
```

```python
# merge = procv
# merge(nome_tabela_onde_procurar, left_on = "nome da coluna onde queremos associar", 
# ... right_on=" nome da coluna na tabela onde procurar, how= "lef or right, usar para preencher qual")
df_principal = df_principal.merge(df_total_de_acoes, left_on= "Ativo",right_on= "Código", how = "left")
df_principal
#OBS, ele ira associar mas trara a coluna rigth junto, necessitando dropar para nao repetir info

# Renomear "Qtde. Teórica"
df_principal = df_principal.rename(columns= {"Qtde. Teórica": "quantidade_teorica"}).copy()
```

```python
# drop = apagar 
df_principal = df_principal.drop(columns= ["Código"])
df_principal
```

```python
# coluna variaçao

df_principal["variacao"] = (df_principal["valor_final"] - df_principal["valor_inicial"]) * df_principal["quantidade_teorica"]

df_principal
```

```python
# formatar numeros float, sumir com notaçao cientifica
pd.options.display.float_format= "{:.2f}".format
df_principal
```

```python
# converter "Qtde. Teórica" para int 
df_principal["quantidade_teorica"] = df_principal["quantidade_teorica"].astype(int)
df_principal
```

```python
# cria coluna subiu pu desceu
# apply, funçao para aplicar outras funçoes, lambda: aplicar em todas linhas
df_principal["Resultado"] = df_principal["variacao"].apply(lambda x:"Subiu" if x>0 else ("Desceu" if x< 0 else "Estavel"))

df_principal
```

```python
# procV nome ,segmento e idade 

df_principal = df_principal.merge(df_Ticker, left_on="Ativo", right_on="Ticker", how= "left")

df_principal = df_principal.merge(df_Chat_GPT, left_on= "Nome",right_on= "Empresa", how = "left")

df_principal
```

```python
df_principal["Faixa etaria"] = df_principal["Idade"].apply(lambda x:"maior que 100" if x > 100 else ("menor que 50" if x < 50 else "entre 50 a 100"))
df_principal
```

```python
# ----------------- Ánalises -------------------------------

maior = df_principal["variacao"].max()

menor = df_principal["variacao"].min()

media = df_principal["variacao"].mean()

media_subiu = df_principal[df_principal["Resultado"]== "Subiu"]["variacao"].mean()

media_desceu = df_principal[df_principal["Resultado"]== "Desceu"]["variacao"].mean()

print(f"Maior \t{maior:,.2f}")
print(f"menor \t{menor:,.2f}")
print(f"media \t{media:,.2f}")
print(f"media_subiu \t{media_subiu:,.2f}")
print(f"media_desceu \t{media_desceu:,.2f}")
```

```python
# criaçao new df com apenas dados que subiram 

df_principal_subiu = df_principal[df_principal["Resultado"]== "Subiu"]
df_principal_subiu
```

```python
# criando df analise segmeto, baseado no df anterior, porem agrupando todos os do mesmo segmento e somando

df_analise_segmento = df_principal_subiu.groupby("Segmento")["variacao"].sum().reset_index() # reset para continuar como dataFrame
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby("Resultado")["variacao"].sum().reset_index() # reset para continuar como dataFrame
df_analise_saldo
```

```python
# graficos
# variavel = px.TipoDoGrafico(tabela, x do grafico, y do grafico, mostrar num do y, titulo do grafico)
fig = px.bar(df_analise_saldo, x="Resultado",y="variacao", text="variacao", title="Valor reais por resultado")
fig.show()
```
